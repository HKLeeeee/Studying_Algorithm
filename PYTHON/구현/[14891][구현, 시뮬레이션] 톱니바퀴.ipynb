```python
def turn_right(gear):
    front = gear[:-1]
    end = gear[-1]
    return end + front

def turn_left(gear):
    front = gear[0]
    end = gear[1:]
    return end + front


def is_same_pole(left, right):
    if left[2] == right[6] :
        return True
    return False

import sys
input = sys.stdin.readline

gears = [None]

for i in range(4) :
    gears.append(input())

# N극 0, S극 1
# 12시 방향부터 차례로, 오른쪽 톱니와는 index 2와 접하고, 왼쪽 톱니와는 index 6과 접함
N = int(input()) # 회전 횟수
for i in range(N) :
    rotation_list = [0, 0, 0, 0, 0]
    idx, rotation = map(int, input().split())
    rotation_list[idx] = rotation

    # check left
    for i in range(idx, 1, -1):
        if is_same_pole(gears[i-1], gears[i]) is False :
            rotation_list[i-1] = -rotation_list[i]

    # check right
    for i in range(idx, 4, 1) :
        if is_same_pole(gears[i], gears[i+1]) is False :
            rotation_list[i+1] = - rotation_list[i]

    # rotation execute
    for i in range(1, 5) :
        if rotation_list[i] == 1 :
            gears[i] = turn_right(gears[i])
        elif rotation_list[i] == -1 :
            gears[i] = turn_left(gears[i])

## scoring
# 2진법
score='0b'
for i in range(4, 0, -1):
    score += gears[i][0]
print(int(score, 2))
```
